```python
#3.tokenization
import regex as re
def tokenize(text):
  return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

```python
# function to remove stopwords
def remove_stop(tokens):
  #pattern = r'\b(?:{})\b'.format('|'.join(stopwords))
  return [t for t in tokens if t.lower() not in stopwords]
```

```python
#pipeline
pipeline = [str.lower, tokenize, remove_stop]
def prepare(text, pipeline):
  tokens = text
  for fun in pipeline:
    tokens = fun(tokens)
  return tokens
```

```python
df['tokens'] = df['Player'].apply(prepare, pipeline=pipeline)
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

```python
#Word frequency Analysis
from collections import Counter
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

```python
freq_df = pd.DataFrame.from_dict(counter,orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

```python
#4.Regex Operation on 5 Random strings
```

```python
import re
#patterns for random strings
regex_patterns = {
    'ip_address': r'^((25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.){3}(25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)$',
    'twitter_username': r'^@[A-Za-z0-9_]{1,15}$',
    'isbn': r'^(97(8|9))?\d{9}(\d|X)$',
    'credit_card': r'^4[0-9]{12}(?:[0-9]{3})?$',
    'time_24h': r'^(?:[01]\d|2[0-3]):[0-5]\d$',
}
#5 Random Strings
Random_strings = {
    'ip_address': '192.168.1.1',
    'twitter_username': '@username123',
    'isbn': '978-3-16-148410-0',
    'credit_card': '4111-1111-1111-1111',
    'time_24h': '13:45',
}

# Function to match strings using regex
def match_strings():
    for pattern_name, pattern in regex_patterns.items():
        regex = re.compile(pattern)
        for string_name, string_value in Random_strings.items():
            if re.match(regex, string_value):
                print(f"'{string_value}' matches the pattern for {pattern_name}.")
            else:
                print(f"'{string_value}' does not match the pattern for {pattern_name}.")

if __name__ == "__main__":
    match_strings()
```

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
comments = [
    "Great player, always performs well under pressure!",
    "Needs to improve batting technique.",
    "Fantastic fielding skills, a real asset to the team.",
    "Inconsistent performance, needs to work harder.",
    "Best captain ever, leads the team with confidence."
]

# Performing sentiment analysis using TextBlob
sentiments = [TextBlob(comment).sentiment.polarity for comment in comments]

# Plot sentiment analysis results
plt.bar(range(len(comments)), sentiments, color='Blue')
plt.xlabel('Comments')
plt.ylabel('Sentiment Polarity')
plt.title('Sentiment Analysis of Comments')
plt.xticks(range(len(comments)), range(1, len(comments) + 1))
plt.show()
```
